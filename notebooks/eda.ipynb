```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path = "../data/raw/home_depot_data_1_2021_12.csv"
```

```python
df = pd.read_csv(data_path)
```

```python
df.info()
```

```python
select_cols = ['title', 'description', 'brand']
```

```python
df[select_cols].head(10)
```

```python
df['title_len'] = df.title.str.len()
df['description_len'] = df.description.str.len()
```

```python
sns.displot(x='title_len', data=df)
plt.title("Title lenght");
```

```python
sns.displot(x='description_len', data=df)
plt.title("Description lenght");
```
